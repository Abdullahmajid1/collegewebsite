```python
import pandas as pd

df = pd.read_csv(r"/content/Assignment1Data_Sample.csv")
print(df.head())
print(df.info())
print(df.describe())
```

```python
#define alll the required colums
imp_fields = ['Object ID', 'Department', 'Object Name', 'Title', 'Culture',
                   'Artist Nationality', 'Object Begin Date', 'Object End Date',
                   'Medium', 'Credit Line', 'Country']
# iterating through all the colums
delete_columns=[colums for colums
                    in df.columns if colums not in imp_fields]
# deleting colums which are not needed
df.drop(columns=delete_columns,inplace=True)
```

```python
#delete rows based on object Id
df.drop_duplicates(subset='Object ID' ,keep='first', inplace=True)
```

```python
# check for duplicates
df['Object ID'].is_unique
```

```python
df
```

```python
unique_values = df['Country'].unique()
print(unique_values)
```

```python
unique_values = df['Culture'].unique()
print(unique_values)
```

```python
#Merging dublicates
df['Country'] = df['Country'].replace('USA', 'United States')
df['Country'] = df['Country'].replace('present-day Uzbekistan', 'Uzbekistan')
df['Country'] = df['Country'].replace('present-day France', 'France')
```

```python
unique_values = df['Country'].unique()
print(unique_values)
```

```python
df
```

```python
#replace missing values
target_column = 'Artist Nationality'
source_column = 'Country'
df[target_column].fillna(df[source_column], inplace=True)
```

```python
df
```

```python
#assigning culture value based on country only which are known
#df.loc[df['Country'] == 'Indonesia', 'Culture'] = 'Asian'
df.loc[df['Country'] == 'Netherlands', 'Culture'] = 'Dutch'
df.loc[df['Country'] == 'United States', 'Culture'] = 'American'
#df.loc[df['Country'] == 'Costa Rica', 'Culture'] = 'Asian'
#df.loc[df['Country'] == 'Papua New Guinea', 'Culture'] = 'Asian'
#df.loc[df['Country'] == 'Peru', 'Culture'] = 'Asian'
df.loc[df['Country'] == 'Egypt', 'Culture'] = 'Middle Eastern'
df.loc[df['Country'] == 'India', 'Culture'] = 'India'
df.loc[df['Country'] == 'Iran', 'Culture'] = 'Iran'
df.loc[df['Country'] == 'France', 'Culture'] = 'French'
#df.loc[df['Country'] == 'Uzbekistan', 'Culture'] = 'Asian'

```

```python
unique_values = df['Culture'].unique()
print(unique_values)
```

```python
#Merging dublicates
df['Culture'] = df['Culture'].replace({'German, Silesia (Bunzlau)': 'German',
                                       'German, Hanau': 'German',
                                       'probably German': 'German',
                                       'possibly Spanis': 'Spanis',
                                       'India (Guler, Punjab Hills)': 'India',
                                       'possibly Greek': 'Greek',
                                       'Minoan, Crete':'Minoan',
                                       'European, Eastern':'European',
                                       'European or Middle Eastern':'Middle Eastern',
                                       'Italian, Venice':'Italian',
                                       'South Netherlandish': 'Dutch',
                                       'British, Stoke-on-Trent, Staffordshire': 'British',
                                       'Byzantine': 'European',
                                       'British, Chelsea': 'British',
                                       'probably French': 'French',
                                       'Romanian':'European',
                                       'possibly Spanish': 'Spanish',
                                       'Greek, Attic': 'Greek',
                                       'Italic':'Italian'

                                       })
```

```python
unique_values = df['Culture'].unique()
print(unique_values)
```

```python
df['Object End Date'] = df['Object End Date'].fillna(0).astype(int)
df['Object Begin Date'] = df['Object Begin Date'].fillna(0).astype(int)
```
